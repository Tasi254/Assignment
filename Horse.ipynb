Advanced prediction for a messy dataset
KNN Imputation is used in this case (K-Nearest Neighbors)

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
# read the data
# The data uses ? for all the missing values
df =pd.read_csv('horse.csv', header=None, na_values= '?')
```

```python
print('The Horse Colic metadata')
print('This data is complex with many missig datas represented by NaN')
print(f'Total columns found: {len(df.columns)}')
print(df.describe()) # this gives the statistical summary metadata
```

```python
#Encoding
print('Data Encoding')
oe= OrdinalEncoder()
df[[1,2]] = oe.fit_transform(df[[1,2]])
print('Feature 1 and 2 are encoded using Ordinal Scaling')
```

```python
#the missing value method (Method: KNN Imputer)
print('The Horse Colic missing data')
# This method is used since more that 30% of the data is missing
# This method checks the 3 most similar horses and uses their data to fill the holes
imputer= KNNImputer(n_neighbors=3)
df_filled = pd.DataFrame(imputer.fit_transform(df))

print ('Imputer is used to intelligently predict missing health data')
print(f'Missing values after KNN: {df_filled.isnull().sum().sum()}')
```
