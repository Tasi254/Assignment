Handling missing data using median Imputation
Handling the outliers in medical health data
This is an online data set anout the Pima Indians Diabetes

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

print("It works!")
```

```python
# Read the data and the Metadata
#Since the data does not have the header, the first thing is to define that
cols =['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Pedigree', 'Age', 'Outcome']
df = pd.read_csv('diabetes.csv', names=cols)

print('Diabetes Data')
print (" Checking the data types")
print(df.dtypes)
print(f'Statement: Data dimensions: {df.shape}')
```

```python
# Encoding the data
print ('Encoding the data')
# Encoding class
le =LabelEncoder()
df['Outcome']= le.fit_transform(df['Outcome'])
print("The numnber of Diabetes is now encoded as the Class Label")
```

```python
# Solving the missing value using the Median Method
print ("Diabetes missing values")
# In this case if we have a 0 in the glucose or the BMI, then that is a missing value/error
#so we proceed to replace 0 with NaN so that Python can  recognise them as missing
df['BMI']= df['BMI'].replace(0, np.nan)
```

```python
#Median is used in this case because medical data usually have some outliers
#That would ruin a simple Mean calculation
median_bmi =df['BMI'].median()
df['BMI']= df['BMI'].fillna(median_bmi)
print(f'Missing BMI values are now replaces by median: {median_bmi}')
```
