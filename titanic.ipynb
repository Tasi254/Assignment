Here mixed data is handles, text and numbers
Simple statistical imputation is used (Mean and Mode)

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#Read data
df =pd.read_csv('titanic.csv')

print ('The Titanic Metadata')
print('Dispaying the structural information')
print(df.info())
print(f'Dataset contains{df.shape[1]} features and {df.shape[0]}records.')
```

```python
#Encoding
print('Encoding the data')
#Encodeing Class (Target)
le =LabelEncoder()
df['Survived'] = le.fit_transform(df['Survived'])
print('The survived class is encoded into 0s and 1s using the LabelEncoder')

#Encoding Feature, the inputs
df = pd.get_dummies(df, columns =['Sex'], prefix ='Feature_Sex')
print('The "sex" feature was One-Hot Encoded to create numeric columns for the model')
```

```python
#Solving for the missing value using the mean and mode
print('The missing data set')
#Mean us the best for Age when data is spread normally
mean_age=df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)
print(f'The missing values are filled with the column mean: {mean_age:.2f}')
```

```python
#Mode is the only way to fill categorical text like Embarked
mode_embarked =df['Embarked'].mode()[0]
df['Embarked']= df['Embarked'].fillna(mode_embarked)
print(f'Missed Embarked values are filled with the most frequent values: {mode_embarked}')
```
